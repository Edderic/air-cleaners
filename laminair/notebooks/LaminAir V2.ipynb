```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fit_factor_grid = pd.read_csv('./laminair/data/LaminAir - v2 fit factor grid.csv')
fit_factor_grid
```

```python
ffg = fit_factor_grid.rename(columns={'x': 'x (inches)'}).set_index('x (inches)')
ffg
```

## Fit Factor Grid

```python
fig, ax = plt.subplots(1,1, figsize=(10,5))
sns.heatmap(ffg, annot=True, ax=ax)
ax.set_xlabel('y (inches)')
ax.set_title('LaminAir V2 Fit Factor Grid')
```
